```python
from sklearn import datasets     #importing datasets from sklearn
import pandas as pd              #importing pandas with an alias pd
import numpy as np               #importing numpy with an alias np 
import matplotlib                #impoting matplotlib
import matplotlib.pyplot as plt  #importing pyplot with an alias plt
#from matplotlib.pyplot import figure #imporiting figure
```

```python
iris = datasets.load_iris()
type(iris)
print(iris.keys())
```

```python
X = iris.data
Y = iris.target
```

```python
df = pd.DataFrame(X, columns = iris.feature_names)
```

```python
colors = ['purple', 'red', 'green']

plt.figure(num=None, figsize=(12,8), dpi=80)

plt.scatter(df['petal length (cm)'], df['petal width (cm)'], c = Y)
''''
cb = plt.colorbar()
loc = np.arange(0,max(Y),max(Y)/float(len(colors)))
cb.set_ticks(loc+0.33)
cb.set_ticklabels(iris.target_names)'''

plt.xlabel('petal length (cm)', fontsize = 15)
plt.ylabel('petal width (cm)', fontsize = 15);
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, 
                                test_size = 0.3, random_state = 21, stratify = Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 6)
knn.fit(X_train, Y_train)
```

```python
knn.predict(X_test)
```

```python
knn.score(X_test, Y_test)
```
